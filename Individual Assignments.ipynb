## 2. What are the top 3 movies with highest nett profit in 2019

```python
# Nomor 2
# Mencari 3 film teratas dengan keuntungan nett terbesar pada tahun 2013

Movies2013 = imdb[imdb.title_year == 2013]
Movies2013.head(10)
data = pd.DataFrame({'Net Profit' : Movies2013['net_USD'], 'judul':Movies2013["movie_title"]})
TopMovies2013 = data.sort_values(by="Net Profit",ascending=False)
TopMovies2013.head(3)

```

## 7. What are the averag budget of all movies in 2013
